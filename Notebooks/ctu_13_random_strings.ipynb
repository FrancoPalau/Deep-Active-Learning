```python
def random_position_string(x):
    maxlen = 10
    if len(x)<=maxlen:
        return x
    else:
        rand_pos = randint(0,len(x)-maxlen)
        return x[rand_pos:rand_pos+maxlen]
```

```python
ctu_13["State"] = ctu_13["State"].apply(random_position_string)
```

```python
def change_class(x):
    if x == "botnet":
        return 1
    elif x== "normal":
        return 0
```

```python
#Varias corridas para la red
for i in range(0,10):
    #read raw dataset
    ctu_13 = pd.read_csv("../Datasets/ctu-13.labeled.cleaned.csv")
    
    #remove first 5 charc.
    ctu_13["State"] = ctu_13["State"].apply(lambda x: x[5:])
    
    #Select sub-string random pos
    ctu_13["State"] = ctu_13["State"].apply(random_position_string)
    
    #clean dataset
    ctu_13["State"] = ctu_13["State"].apply(lambda x: np.nan if x=="" else x)
    ctu_13.rename(columns={"State":"domain"}, inplace=True)
    ctu_13.drop(["src","dst","port","proto","label","modelsize"], axis=1, inplace=True)
    ctu_13["class"] = ctu_13["class"].apply(change_class)
    ctu_13.dropna(inplace=True)
    
    #Save dataset
    name = "../Datasets/Random_pos/ctu_13_without_first_5_charc_and_rand_pos_maxlen10_run"+str(i)+".csv"
    ctu_13.to_csv(name, index=False)
```
